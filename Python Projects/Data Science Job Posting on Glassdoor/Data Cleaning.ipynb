### Some of the objectives we need to achieve in order to make this dataset clean and easy to visualize.

-- It looks like the nulls have been replaced by -1, everywhere in the dataset. So,we will look to replace it with nan.

-- The Company Name column is concatenated with the rating,so we will look into creating a clean column for the Company Name.

-- The Salary Estimate column looks little messed up having a lot of details in a single columns.
        Let's try to break it down to 3 new columns(Min Salary Estimate,
        Max Salary Estimate,Average Salary Estimate).

-- We will look if we can add some useful columns for generating insights using the Job Description column.

-- Add a new column for the age of the company.(Letting current year be 2023)

-- Add a new column specifying if the job posting is related to a Junior level role or a senior level role.

```python
# Replacement of "-1"
for i in data.columns:
    data[i].replace({"-1":np.nan,-1.0:np.nan,-1:np.nan},inplace=True)
data
```

##### Company Name column rectification

```python
zeroth=lambda i:i[0]
data["Company Name"]=data["Company Name"].str.split("\n").map(zeroth)
data.head()
```

```python
#new columns using salary estimate column 
data["Salary Estimate"]=data["Salary Estimate"].str.split("(").map(zeroth)
data.head()
```

```python
data["Min Salary Estimate"]=data["Salary Estimate"].str.split("-").map(zeroth).str.replace("$","",regex=False).str.replace("K","",regex=False).astype("int")
data.head()
```

```python
data["Max Salary Estimate"]=data["Salary Estimate"].str.split("-").map(lambda i:i[1]).str.replace("$","",regex=False).str.replace("K","",regex=False).astype("int")
data.head()
```

```python
data["Avg Salary Estimate"]=(data["Max Salary Estimate"]+data["Min Salary Estimate"])/2
data.head()
```

#### Generating new columns using job description

```python
data["Job Description"].loc[0].lower().split("\n")
```

##### Looks like the description includes the important skills required for the specific job opening. We could use this to generate important insights

```python
#essential skill
lst=['python','excel','hadoop','spark','aws','tableau','big data','sql','natural language processing',
     "machine learning",'deep learning','numpy','pandas','matplotlib','seaborn','sklearn','scipy']
for i in lst:
    data[i]=data["Job Description"].str.lower().str.contains(i).astype("int")
```

```python
# Adding a new age column
data["Age"]=2023-data["Founded"]
data.head()
```
